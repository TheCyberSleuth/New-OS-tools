```python
import pandas as pd
import folium
from IPython.display import display

# Load data
file_path = "kerala_crime_dummy_data.csv"
data = pd.read_csv(file_path)

# Define color mapping for crime types
crime_colors = {
    "Theft": "blue",
    "Assault": "red",
    "Fraud": "purple",
    "Robbery": "orange",
    "Murder": "darkred",
    "Kidnapping": "green",
    "Cyber Crime": "cadetblue",
    # Add more crime types as needed
}



map_center = [data["Latitude"].mean(), data["Longitude"].mean()]
crime_map = folium.Map(location=map_center, zoom_start=7)

 Add markers with color based on crime type
for _, row in data.iterrows():
    crime_type = row["Crime_Type"]
    marker_color = crime_colors.get(crime_type, "gray")  # Default to gray if not in dict
    
    folium.Marker(
        location=[row["Latitude"], row["Longitude"]],
        popup=f"{row['District']} - {crime_type}<br>Date: {row['Date']}<br>Status: {row['Status']}",
        tooltip=crime_type,
        icon=folium.Icon(color=marker_color)
    ).add_to(crime_map)

# Display in Jupyter
display(crime_map)
```

```python
import pandas as pd
import folium
from IPython.display import display

# Load the dataset
file_path = "kerala_crime_dummy_data.csv"
data = pd.read_csv(file_path)

# Define mapping of numbers to crime types and colors
crime_options = {
    1: ("Theft", "blue"),
    2: ("Assault", "red"),
    3: ("Fraud", "purple"),
    4: ("Robbery", "orange"),
    5: ("Murder", "darkred"),
    6: ("Kidnapping", "green"),
    7: ("Cyber Crime", "cadetblue")
}

# Show options to user
print("Select Crime Type to Display on Map:")
for key, value in crime_options.items():
    print(f"{key}: {value[0]}")

# Get user input
choice = int(input("Enter a number: "))
if choice not in crime_options:
    print("❌ Invalid choice.")
else:
    selected_crime, color = crime_options[choice]
    
    # Filter data for selected crime type
    filtered_data = data[data["Crime_Type"] == selected_crime]
    
    if filtered_data.empty:
        print(f"No records found for {selected_crime}.")
    else:
        # Create map centered on Kerala
        map_center = [filtered_data["Latitude"].mean(), filtered_data["Longitude"].mean()]
        crime_map = folium.Map(location=map_center, zoom_start=7)
        
        # Add markers for selected crime
        for _, row in filtered_data.iterrows():
            folium.Marker(
                location=[row["Latitude"], row["Longitude"]],
                popup=f"{row['District']} - {row['Crime_Type']}<br>Date: {row['Date']}<br>Status: {row['Status']}",
                tooltip=row["Crime_Type"],
                icon=folium.Icon(color=color)
            ).add_to(crime_map)
        
        # Display in Jupyter
        display(crime_map)
```

```python
class Animal:
    def __init__(self, name, age, sound):
        self.name = name
        self.age = age
        self.sound = sound
    def make_sound(self):
        print(self.sound)

class Dog(Animal):
    def __init__(self, name, age, sound):
        super().__init__(name, age, sound)
        
    def Action(self):
        print("moves tails")

class Cat(Animal):
    def __init__(self, name, age, sound):
        super().__init__(name, age, sound)
    def Action(self):
        print("moves ears")

dog = Dog("Buddy", 3, "Woof")
cat = Cat("Whiskers", 2, "Meow")

dog.Action()
dog.make_sound()
cat.Action()
cat.make_sound()
```

```python
class Animal:
    def __init__(self, name, age, sound, breed=None): 
        self.name = name
        self.age = age
        self.sound = sound
        self.breed = breed

    def make_sound(self):
        print(self.sound)

class Dog(Animal):
    def __init__(self, name, age, sound, breed):
        super().__init__(name, age, sound, breed)

    def Action(self):
        print("Moves tail")

    def Breed(self):
        print("Breed of dog is " + self.breed)

class Cat(Animal):
    def __init__(self, name, age, sound):
        super().__init__(name, age, sound)

    def Action(self):
        print("Moves ears")

class Bird(Animal):
    def __init__(self, name, age, sound):
        super().__init__(name, age, sound)

    def Fly(self):
        print("Flies in the sky")

dog = Dog("Buddy", 3, "Woof", "Poodle")
cat = Cat("Whiskers", 2, "Meow")

dog.Breed()
```
